```python
# Gln or Glu / Z	CAA, CAG; GAA, GAG
# Asn or Asp / B	AAU, AAC; GAU, GAC
#Asn or Asp / B	AAU, AAC; GAU, GAC
# START	AUG

codon = """Ala / A	GCU, GCC, GCA, GCG
Ile / I	AUU, AUC, AUA
Arg / R	CGU, CGC, CGA, CGG; AGA, AGG
Leu / L	CUU, CUC, CUA, CUG; UUA, UUG
Asn / N	AAU, AAC
Lys / K	AAA, AAG
Asp / D	GAU, GAC
Met / M	AUG
Phe / F	UUU, UUC
Cys / C	UGU, UGC
Pro / P	CCU, CCC, CCA, CCG
Gln / Q	CAA, CAG
Ser / S	UCU, UCC, UCA, UCG; AGU, AGC
Glu / E	GAA, GAG
Thr / T	ACU, ACC, ACA, ACG
Trp / W	UGG
Gly / G	GGU, GGC, GGA, GGG
Tyr / Y	UAU, UAC
His / H	CAU, CAC
Val / V	GUU, GUC, GUA, GUG
STOP	UAA, UGA, UAG""".strip()
```

```python
code = {}
for each in codon.split("\n"):
    k,v = each.split("\x09")
    if '/' in k:
        k = k.split('/')[-1]
    v = v.replace(",", "").replace(";", "").lower().replace("u", "t").split(" ")
    for vv in v:
        code[vv] = k
#code 
```

```python
amino_acid = []
for r in range(3):
    for i in range(r, len(genome)-3, 3):
        amino_acid.append(code[genome[i:i+3]])

amino_acid = ''.join(amino_acid).replace(' ', '')
```

```python
peptides = []
for each in amino_acid.split('STOP'):
    for i in each.split("M", 1):
        peptides.append("M"+ i)
```
